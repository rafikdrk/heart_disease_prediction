## Modeling

### split the data

```python

train,test=train_test_split(data,random_state=123,test_size=.2)
# Define the target variable
y_train = train['heartdisease'].values
y_test = test['heartdisease'].values

# Remove the 'heartdisease' column from the feature sets
x_train = train
x_test = test
del x_train['heartdisease']
del x_test['heartdisease']
```

```python
dv=DictVectorizer(sparse=False)
x_train=x_train.to_dict(orient='records')
x_test=x_test.to_dict(orient='records')

x_train=dv.fit_transform(x_train)
x_test=dv.transform(x_test)
```

### logistic regression

```python
# Create a Logistic Regression model
lr = LogisticRegression(C=1,max_iter=2000)

# Fit the model to the training data
lr.fit(x_train, y_train)

# Make predictions on the test data
y_pred = lr.predict(x_test)
```

```python
# Evaluate the model
lr.score(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

### AdaBoost

```python
# Create a AdaBoost model
ada = AdaBoostClassifier(n_estimators=6,learning_rate=0.3)
ada.fit(x_train, y_train)
y_pred = ada.predict(x_test)
```

```python
# Evaluate the model
# train accuracy
ada.score(x_train,y_train)
```

```python
# test accuracy
ada.score(x_test,y_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

### XGBoost

```python
xgb = XGBClassifier(n_estimators=10,max_depth=5, learning_rate=0.2)
xgb.fit(x_train, y_train)
y_pred = xgb.predict(x_test)
```

```python
xgb.score(x_train,y_train)
```

```python
xgb.score(x_test,y_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

### Random forest

```python
rf=RandomForestClassifier(max_depth=15,n_estimators=12)
rf.fit(x_train,y_train)
y_pred = rf.predict(x_test)
```

```python
rf.score(x_train,y_train)
```

```python
rf.score(x_test,y_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

### Decision Tree

```python
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
y_pred = dt.predict(x_test)
```

```python
dt.score(x_train,y_train)
```

```python
dt.score(x_test,y_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

It looks like that the best model based on the accuracy is XGBoost by 85.5% accuracy

### Best model

```python
xgb = XGBClassifier(n_estimators=10,max_depth=5, learning_rate=0.2)
xgb.fit(x_train, y_train)
y_pred = xgb.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
mean_absolute_error(y_test, y_pred)
```
